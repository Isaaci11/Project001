## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style='whitegrid')
print('Libraries loaded successfully.')
```

## 2. Load and Inspect Data

Replace the sample data below with your own dataset (e.g. `pd.read_csv('your_file.csv')`).

```python
# ── Sample dataset ────────────────────────────────────────────────────────────
np.random.seed(42)
n = 200

df = pd.DataFrame({
    'age':     np.random.randint(18, 65, n),
    'income':  np.random.normal(55000, 15000, n).round(2),
    'score':   np.random.uniform(0, 100, n).round(1),
    'group':   np.random.choice(['A', 'B', 'C'], n),
})

# Introduce a few missing values for demonstration
df.loc[np.random.choice(df.index, 10, replace=False), 'income'] = np.nan

print(f'Shape: {df.shape}')
df.head()
```

## 3. Clean and Preprocess Data

```python
print('Missing values per column:')
print(df.isnull().sum())

# Fill missing income values with the column median
df['income'] = df['income'].fillna(df['income'].median())

print('\nMissing values after cleaning:')
print(df.isnull().sum())
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f'Duplicate rows: {duplicates}')
if duplicates > 0:
    df = df.drop_duplicates()
    print(f'Shape after removing duplicates: {df.shape}')
```

## 4. Exploratory Data Analysis

```python
df.describe()
```

```python
print('Group counts:')
print(df['group'].value_counts())

print('\nMean score by group:')
print(df.groupby('group')['score'].mean().round(2))
```

## 5. Visualizations

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 9))
fig.suptitle('Exploratory Data Analysis', fontsize=14)

# Distribution of Age
axes[0, 0].hist(df['age'], bins=20, color='steelblue', edgecolor='white')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Count')

# Distribution of Income
axes[0, 1].hist(df['income'], bins=20, color='coral', edgecolor='white')
axes[0, 1].set_title('Income Distribution')
axes[0, 1].set_xlabel('Income ($)')
axes[0, 1].set_ylabel('Count')

# Score by Group (box plot)
df.boxplot(column='score', by='group', ax=axes[1, 0], grid=False)
axes[1, 0].set_title('Score by Group')
axes[1, 0].set_xlabel('Group')
axes[1, 0].set_ylabel('Score')
plt.sca(axes[1, 0])          # suppress the automatic suptitle from boxplot
plt.title('Score by Group')

# Age vs Income (scatter)
scatter = axes[1, 1].scatter(
    df['age'], df['income'],
    c=df['score'], cmap='viridis', alpha=0.6, edgecolors='none'
)
plt.colorbar(scatter, ax=axes[1, 1], label='Score')
axes[1, 1].set_title('Age vs Income (colored by Score)')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Income ($)')

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6, 4))
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 6. Summary Statistics

```python
summary = pd.DataFrame({
    'mean':   numeric_cols.mean(),
    'median': numeric_cols.median(),
    'std':    numeric_cols.std(),
    'min':    numeric_cols.min(),
    'max':    numeric_cols.max(),
}).round(2)

print('Summary statistics:')
summary
```
